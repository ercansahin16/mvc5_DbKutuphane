# [dbo].[EnAktifUye]

```sql
/****** Object:  StoredProcedure [dbo].[EnAktifUye]    Script Date: 1.08.2024 11:12:12 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[EnAktifUye]
AS
SELECT TOP 1 TBLUYELER.ADI+TBLUYELER.SOYADI FROM TBLHAREKET
INNER JOIN TBLUYELER ON TBLUYELER.ID=TBLHAREKET.UYE
GROUP BY TBLUYELER.ADI,TBLUYELER.SOYADI ORDER BY COUNT(*) DESC
GO
```

# [dbo].[EnBasariliPersonel]

```sql
/****** Object:  StoredProcedure [dbo].[EnBasariliPersonel]    Script Date: 1.08.2024 11:12:12 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[EnBasariliPersonel]
AS
SELECT TOP 1 TBLPERSONEL.PERSONEL FROM TBLHAREKET
INNER JOIN TBLPERSONEL ON TBLPERSONEL.ID=TBLHAREKET.PERSONEL 
WHERE  TBLHAREKET.ISLEMDURUM = 1
GROUP BY TBLPERSONEL.PERSONEL  ORDER BY COUNT(*) DESC 
GO
```
